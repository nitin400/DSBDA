```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
```

```python
soup
```

```python
#fetching the reviews /comments
reviews=soup.findAll('div',{'class':'t-ZTKy'});
for review in reviews:
    print(review.get_text()+"\n\n\n")
```

```python
#Average overall rating
ratings=soup.find('div',{'class':'_2d4LTz'}).get_text();
print(ratings)
```

```python
#individual Ratings
individual_ratings=soup.findAll('div',{'class':'_3LWZlK _1BLPMq'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text()+ "\n")
```

```python
tags=soup.find('span',{'class':'_3Ay6Sb _31Dcoz'}).get_text();
tags
```

```python
#customer_names
cust_name=soup.findAll('p',{'class':'_2mcZGG'})
for i in cust_name:
    print(i.get_text());
    
```
